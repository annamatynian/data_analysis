# MEDIAN



```python
#Median and mean of Wins (W)
print(df.W.mean())
df.W.median()
```

```python
print(df.W.median())
```

```python
#What is the median of wins of the Baltimore team (BAL) until 2000

df.loc[(df.Team == "BAL") & (df.Year < 2000)].W.median()
```

# MODE



```python
# The mode, median and mean of OBP
print(df.OBP.mean())
print(df.OBP.median())
df.OBP.mode()
```

```python
#Mode of OBP during the year 2010
df.loc[df.Year == 2010].OBP.mode()
```

# CORRELATION


```python
#Pick variables
df_correlation = df[["RS", "RA", "W"]]
df_correlation.head(1)
```

```python
df_correlation.corr()
```

```python
#Correlation heatmap
sns.heatmap(data = df_correlation.corr()
, annot = True, fmt = ".2g", center = 0, cmap = "coolwarm", linewidth = 1, linecolor = "black")
```

```python
df_correlation2 = df[["OBP", "SLG", "BA"]]
df_correlation2.head()
```

```python
df_correlation2.corr()
```

```python
sns.heatmap(data = df_correlation2.corr(), annot = True, fmt = ".1g", center = 0.7, cmap = "rocket")
```
