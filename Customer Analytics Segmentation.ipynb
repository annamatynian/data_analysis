```python
# Load the data, contained in the segmentation data csv file.
df_segmentation = pd.read_csv("D:\PYTHON LAB\segmentation data.csv", index_col = 0)
```

## ${\textbf{Correlation Estimate}}$

```python
# Compute Pearson correlation coefficient for the features in our data set.
#E.g. the correlation coefficient 0.654605 is showing us that there is a strong correlation between age and education
df_segmentation.corr()
```

```python
#Visualizing the correlation between the variables

plt.figure(figsize = (12, 9))
s = sns.heatmap(df_segmentation.corr(),
               annot = True, 
               cmap = 'viridis',
               vmin = -1, 
               vmax = 1)
s.set_yticklabels(s.get_yticklabels(), rotation = 0, fontsize = 12)
s.set_xticklabels(s.get_xticklabels(), rotation = 90, fontsize = 12)
plt.title('Correlation Heatmap')
plt.show()

# The yellow segments of the heatmaps are displaying the highest positive correlation between the variables 
# Dark-blue segments are showing the lowest negative correlation
```

## ${\textbf{Visualize Raw Data}}$

```python
# Plotting the data. Create a 12 by 9 inches figure.
# We have 2000 data points, which we'll scatter across Age and Income, located on positions 2 and 4 in our data set. 

plt.figure(figsize = (12, 9))
plt.scatter(df_segmentation.iloc[:, 2], df_segmentation.iloc[:, 4])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Visualization of raw data')
```
